```python
# 3.2 Handle missing values
print("\nMissing Values Analysis:")
display(df_clean.isnull().sum())

# 1. For Offense End DateTime
if 'Offense End DateTime' in df_clean.columns and 'Offense Start DateTime' in df_clean.columns:
    # Create a new column that indicates if end time is missing
    df_clean['end_time_missing'] = df_clean['Offense End DateTime'].isnull()

    # Fill missing end times with start times + 1 hour (a reasonable assumption)
    end_time_missing = df_clean['Offense End DateTime'].isnull()
    if end_time_missing.any():
        df_clean.loc[end_time_missing, 'Offense End DateTime'] = df_clean.loc[end_time_missing, 'Offense Start DateTime'] + pd.Timedelta(hours=1)
        print(f"Added estimated end time for {end_time_missing.sum()} incidents")

# 2. For 100 Block Address
if '100 Block Address' in df_clean.columns:
    # Create a flag for missing addresses
    df_clean['address_missing'] = df_clean['100 Block Address'].isnull()

    # Create a generic address label for missing values
    address_missing = df_clean['100 Block Address'].isnull()
    if address_missing.any():
        df_clean.loc[address_missing, '100 Block Address'] = "LOCATION UNSPECIFIED"
        print(f"Added placeholder for {address_missing.sum()} missing addresses")

# 3. For Missing Coordinates
if 'Latitude' in df_clean.columns and 'Longitude' in df_clean.columns:
    # Flag records with missing coordinates
    df_clean['coords_missing'] = df_clean['Latitude'].isnull() | df_clean['Longitude'].isnull()

    # For spatial analysis, we'll still need to filter these out when doing geo calculations
    # but we'll keep them in the dataset for other analyses
    coords_missing = df_clean['coords_missing'].sum()
    print(f"Flagged {coords_missing} records with missing coordinates")

    # For records that will be used in spatial analysis, ensure they have coordinates
    spatial_df = df_clean[~df_clean['coords_missing']].copy()
    print(f"Created spatial subset with {len(spatial_df)} complete records for geo analysis")

# 4. For other columns with few missing values
for col in df_clean.columns:
    missing = df_clean[col].isnull().sum()
    if 0 < missing <= 10 and col not in ['Latitude', 'Longitude', 'Offense End DateTime', '100 Block Address']:
        # Create a missing flag column
        df_clean[f'{col}_missing'] = df_clean[col].isnull()

        if df_clean[col].dtype == 'object':
            # For categorical, use a special category
            df_clean[col].fillna('UNKNOWN', inplace=True)
        else:
            # For numeric columns, use a special value like -999 (often used as a missing indicator)
            df_clean[col].fillna(-999, inplace=True)

        print(f"Marked and filled {missing} missing values in '{col}'")

# Create a summarizing column that counts how many values are missing in each row
missing_columns = [col for col in df_clean.columns if col.endswith('_missing')]
if missing_columns:
    df_clean['total_missing_count'] = df_clean[missing_columns].sum(axis=1)
    print(f"Added total missing count column (max missing per record: {df_clean['total_missing_count'].max()})")

print("\nMissing Value Handling Complete")
```

```python
# 4.1 Create a geodataframe for spatial operations
import geopandas as gpd
from shapely.geometry import Point

# Convert to GeoDataFrame
geometry = [Point(xy) for xy in zip(df_clean['Longitude'], df_clean['Latitude'])]
geo_df = gpd.GeoDataFrame(df_clean, geometry=geometry, crs="EPSG:4326")

print("\nGeoDataFrame created successfully")
print(f"Number of points: {len(geo_df)}")
```

```python
import numpy as np
import pandas as pd

def create_safety_grid(geo_df, initial_cell_size=0.05, max_cells=1_000_000, adjust_cell_size=True):
    """
    Create a spatial grid covering geo_df's extent with adaptive cell sizing to prevent memory overload.
    Args:
        geo_df: GeoDataFrame with geometries.
        initial_cell_size: Starting size of each cell (in degrees).
        max_cells: Maximum number of cells to prevent memory issues.
        adjust_cell_size: Automatically adjust cell_size if too many cells are estimated.
    Returns:
        grid_df: DataFrame with grid cells and metadata.
    """
    x_min, y_min, x_max, y_max = geo_df.total_bounds
    print(f"Bounds: x_min={x_min}, x_max={x_max}, y_min={y_min}, y_max={y_max}")

    cell_size = initial_cell_size
    attempt = 1

    # Function to estimate total cells based on current cell size
    def estimate_total_cells(cell_size):
        x_coords = np.arange(x_min, x_max, cell_size)
        y_coords = np.arange(y_min, y_max, cell_size)
        return len(x_coords) * len(y_coords), x_coords, y_coords

    total_cells, x_coords, y_coords = estimate_total_cells(cell_size)
    print(f"Attempt {attempt}: Cell size = {cell_size}, Estimated total cells = {total_cells}")

    # Auto-adjust cell size to reduce number of cells if necessary
    while adjust_cell_size and total_cells > max_cells:
        cell_size *= 2  # Double the cell size to reduce number of cells
        total_cells, x_coords, y_coords = estimate_total_cells(cell_size)
        attempt += 1
        print(f"Attempt {attempt}: Increased cell size to {cell_size}, Estimated total cells = {total_cells}")

    # If still too large, raise an error
    if total_cells > max_cells:
        raise MemoryError(f"Even after adjusting, attempting to create {total_cells} cells exceeds limit {max_cells}. "
                          f"Consider setting a higher initial cell size.")

    # Now create the cells
    cells = []
    for x in x_coords:
        for y in y_coords:
            cell = {
                'grid_id': f"{x:.6f}_{y:.6f}",
                'cell_x': x,
                'cell_y': y,
                'cell_x_max': x + cell_size,
                'cell_y_max': y + cell_size,
                'cell_center_x': x + cell_size / 2,
                'cell_center_y': y + cell_size / 2,
                'incident_count': 0,
                'avg_severity': 0,
                'latest_date': None
            }
            cells.append(cell)

    grid_df = pd.DataFrame(cells)
    print(f" Created grid with {len(grid_df)} cells using cell size {cell_size}")
    return grid_df
```

```python
# 1. Create the grid from your function (if not already created)
grid_df = create_safety_grid(geo_df, initial_cell_size=0.01)

# 2. Convert the grid dataframe to a geodataframe with polygon geometries
from shapely.geometry import box

# Create polygon geometries for each grid cell
grid_geometries = [box(row.cell_x, row.cell_y, row.cell_x_max, row.cell_y_max)
                  for _, row in grid_df.iterrows()]

# Create the grid geodataframe
grid_gdf = gpd.GeoDataFrame(grid_df, geometry=grid_geometries, crs=geo_df.crs)

# 3. Perform the spatial join between incidents (points) and grid cells (polygons)
joined = gpd.sjoin(geo_df, grid_gdf, how='inner', predicate='within')

# 4. Aggregation (incident count and latest date)
agg_results = joined.groupby('grid_id').agg(
    incident_count=('grid_id', 'size'),  # Count of incidents
    latest_date=('Offense Start DateTime', 'max')  # Latest offense date
).reset_index()

print("Aggregated results preview:\n", agg_results.head())

# 5. Merge with grid, but rename columns in agg_results to avoid conflicts
grid_gdf = grid_gdf.merge(
    agg_results.rename(columns={'incident_count': 'incident_count_agg', 'latest_date': 'latest_date_agg'}),
    on='grid_id',
    how='left'
)

# 6. Update existing columns with aggregated values
grid_gdf['incident_count'] = grid_gdf['incident_count_agg'].fillna(0).astype(int)
grid_gdf['latest_date'] = grid_gdf['latest_date_agg'].fillna(grid_gdf['latest_date'])

# 7. Drop temporary aggregated columns
grid_gdf = grid_gdf.drop(columns=['incident_count_agg', 'latest_date_agg'])

# 8. Filter non-empty cells
grid_gdf = grid_gdf[grid_gdf['incident_count'] > 0]

print(f" Final grid cells with incidents: {len(grid_gdf)}")

# 9. Calculate safety score
# Normalize incident count (higher count = lower safety)
if len(grid_gdf) > 0:
    max_count = grid_gdf['incident_count'].max()
    if max_count > 0:
        grid_gdf['incident_score'] = 1 - (grid_gdf['incident_count'] / max_count)
    else:
        grid_gdf['incident_score'] = 1.0

    # Normalize severity (higher severity = lower safety)
    max_severity = grid_gdf['avg_severity'].max()
    if max_severity > 0:
        grid_gdf['severity_score'] = 1 - (grid_gdf['avg_severity'] / max_severity)
    else:
        grid_gdf['severity_score'] = 1.0

    # Calculate final safety score (weighted average)
    grid_gdf['safety_score'] = (0.6 * grid_gdf['incident_score'] + 0.4 * grid_gdf['severity_score'])

    # Scale to 0-100 for easier interpretation
    grid_gdf['safety_score'] = grid_gdf['safety_score'] * 100

# Display safety grid sample
print("\nSafety Grid Sample:")
display(grid_gdf.head())
```

```python
# Create time-based safety scores
# For each grid cell, analyze safety by time of day
time_periods = [
    (0, 6, 'Night (12AM-6AM)'),
    (6, 12, 'Morning (6AM-12PM)'),
    (12, 18, 'Afternoon (12PM-6PM)'),
    (18, 24, 'Evening (6PM-12AM)')
]
```

```python
import numpy as np
import pandas as pd

def time_based_safety_fast(grid_df, incidents_df, time_periods):
    """
    Optimized function to calculate time-based safety scores without nested loops.
    """

    # 1. Assign each incident to a grid cell using np.digitize (fast binning)
    x_bins = np.sort(grid_df['cell_x'].unique())
    y_bins = np.sort(grid_df['cell_y'].unique())

    # Find the bin index for each incident (right bin edge)
    incidents_df['x_bin'] = np.digitize(incidents_df['Longitude'], x_bins) - 1
    incidents_df['y_bin'] = np.digitize(incidents_df['Latitude'], y_bins) - 1

    # Map bin index back to grid_id
    grid_df['bin_x'] = grid_df['cell_x'].rank(method='dense').astype(int) - 1
    grid_df['bin_y'] = grid_df['cell_y'].rank(method='dense').astype(int) - 1

    grid_map = grid_df.set_index(['bin_x', 'bin_y'])['grid_id'].to_dict()

    # Assign grid_id to each incident (drop if outside bounds)
    incidents_df['grid_id'] = incidents_df.set_index(['x_bin', 'y_bin']).index.map(grid_map)
    incidents_df = incidents_df.dropna(subset=['grid_id'])  # Drop incidents outside the grid

    # 2. Explode incidents over time periods
    time_data = []
    for start_hour, end_hour, period_name in time_periods:
        period_incidents = incidents_df[
            (incidents_df['Hour'] >= start_hour) & (incidents_df['Hour'] < end_hour)
        ].copy()
        period_incidents['time_period'] = period_name
        time_data.append(period_incidents)

    # Combine all time periods back into a single DataFrame
    all_period_incidents = pd.concat(time_data, ignore_index=True)

    # 3. Group by grid and time period to calculate stats
    grouped = all_period_incidents.groupby(['grid_id', 'time_period']).agg(
        incident_count=('grid_id', 'count'),
        avg_severity=('Severity', 'mean')  # If 'Severity' missing, handle separately later
    ).reset_index()

    # 4. Add center_x and center_y to result
    center_coords = grid_df[['grid_id', 'cell_center_x', 'cell_center_y']].set_index('grid_id')
    grouped = grouped.join(center_coords, on='grid_id')

    # 5. Normalize counts and severity within each time period
    grouped['incident_score'] = 0.0
    grouped['severity_score'] = 0.0

    for period_name in [p[2] for p in time_periods]:
        period_mask = grouped['time_period'] == period_name
        period_df = grouped[period_mask]

        # Normalize incident count
        max_count = period_df['incident_count'].max()
        if max_count > 0:
            grouped.loc[period_mask, 'incident_score'] = 1 - (period_df['incident_count'] / max_count)
        else:
            grouped.loc[period_mask, 'incident_score'] = 1.0  # Perfectly safe if no incidents

        # Normalize severity
        if 'avg_severity' in period_df.columns:
            max_severity = period_df['avg_severity'].max()
            if max_severity > 0:
                grouped.loc[period_mask, 'severity_score'] = 1 - (period_df['avg_severity'] / max_severity)
            else:
                grouped.loc[period_mask, 'severity_score'] = 1.0  # Safe if no severity reported

    # 6. Compute final safety score (weighted average)
    grouped['safety_score'] = grouped['incident_score'] * 0.6 + grouped['severity_score'] * 0.4
    grouped['safety_score'] = grouped['safety_score'] * 100  # Scale 0-100

    return grouped
```

```python
# Run optimized function
time_safety_df = time_based_safety_fast(grid_df, df_clean, time_periods)

# Display sample
print("\nTime-based Safety Sample (Optimized):")
display(time_safety_df.head())
```

```python
from folium.plugins import MarkerCluster

# Create incident map with cluster markers
def create_incident_map(df, center_lat=None, center_lon=None):
    """Create an interactive map of incidents with clustering"""
    # If center coordinates aren't provided, calculate them from the data
    if center_lat is None or center_lon is None:
        center_lat = df['Latitude'].mean()
        center_lon = df['Longitude'].mean()

    # Create a base map
    m = folium.Map(location=[center_lat, center_lon], zoom_start=12)

    # Create a marker cluster
    marker_cluster = MarkerCluster().add_to(m)

    # Add markers for each incident
    for idx, row in df.sample(min(1000, len(df))).iterrows():  # Sample to avoid overloading the map
        # Create popup content
        popup_content = f"""
        <b>Offense:</b> {row.get('Offense', 'Unknown')}<br>
        <b>Date:</b> {row.get('Offense Start DateTime', 'Unknown')}<br>
        <b>Location:</b> {row.get('100 Block Address', 'Unknown')}<br>
        <b>Precinct:</b> {row.get('Precinct', 'Unknown')}
        """

        # Add marker to cluster
        folium.Marker(
            location=[row['Latitude'], row['Longitude']],
            popup=folium.Popup(popup_content, max_width=300),
            icon=folium.Icon(color='red', icon='info-sign')
        ).add_to(marker_cluster)

    return m

# Calculate center coordinates from the data
center_lat = df_clean['Latitude'].mean()
center_lon = df_clean['Longitude'].mean()

# Create incident map
incident_map = create_incident_map(df_clean, center_lat, center_lon)

# Save incident map to HTML file
incident_map.save('incident_map.html')
print("Incident map created and saved as 'incident_map.html'")
```
